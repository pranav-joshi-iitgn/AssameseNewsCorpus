```python
base_url = "https://www.dainandinbartagroup.in/homepage/"
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from numpy import unique
c1 = "listing-item-thumbnail"
c2 = "sidebar-column"
```

```python
driver = webdriver.Firefox()
driver.get(base_url)
SB = driver.find_element(By.CLASS_NAME,c2)
next_butt = SB.find_element(By.CLASS_NAME,"next")
batch_size = 80
URLS = []
n = 4902
i=0
while i < n:
    try:
        dis = driver.find_elements(By.ID,'dismiss-button')
        if len(dis):
            dis[0].click()
            print("click")
            time.sleep(1)
            SB = driver.find_element(By.CLASS_NAME,c2)
            next_butt = SB.find_element(By.CLASS_NAME,"next")
        D = SB.find_elements(By.CLASS_NAME,c1)
        D = [x.find_element(By.TAG_NAME,"a") for x in D]
        D = [x.get_attribute("href") for x in D]
        URLS.extend(D)
        next_butt.click()
        if i % batch_size == batch_size-1:
            f = open(f"URLS3/urls_{i//batch_size}.txt",'w')
            f.write("\n".join(URLS))
            f.close()
            URLS = []
        i += 1
    except:
        pass
driver.close()
```

```python
S = []
for i in range(62):
    S.extend(open(f"URLS3/urls_{i}.txt",'r').read().split("\n"))
S = unique(S)
f = open("fullurls3.txt",'w')
f.write("\n".join(S))
f.close()
```

```python
n = len(S)
n = n//5
for i in range(6):
    f = open(f"fullurls3_block_{i}.txt",'w')
    f.write("\n".join(S[n*i:n*(i+1)]))
    f.close()
```
