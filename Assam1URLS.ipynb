````python
driver = webdriver.Firefox()
driver.get(base_url)
time.sleep(3)
dismiss = driver.find_elements(By.ID,"dismiss-button")
if len(dismiss) > 0:
    dismiss[0].click()
    print("dismissed")
    time.sleep(1)
else:
    print("no ad")
load_more = driver.find_elements(By.ID,'gmag-free-load-more-link')[0]

for i in range(5):
    load_more.click()
    time.sleep(2)
Arts = driver.find_elements(By.TAG_NAME,'article')
urls = []
for art in Arts:
    D = art.find_elements(By.CLASS_NAME,'entry-header')
    assert len(D)==1
    D = D[0]
    H = D.find_elements(By.TAG_NAME,'h2')[0]
    A = H.find_elements(By.TAG_NAME,'a')[0]
    urls.append(A.get_attribute("href"))
driver.close()
f = open("URLS1.txt",'w')
f.write("\n".join(urls))
f.close()
urls
````
