```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
base_url = "https://www.bbc.com/business/technology-of-business"
```

### URL extraction

```python
driver = webdriver.Firefox()
driver.get(base_url)
link = []
next_buttons = driver.find_elements(By.CSS_SELECTOR,'[data-testid="pagination-next-button"]')
next_buttons[0].click() #to next page
div1 = driver.find_elements(By.CSS_SELECTOR, '[data-testid="alaska-section"]')
div1 = div1[0]
tag_a = div1.find_elements(By.TAG_NAME, 'a')
for i in range(len(tag_a)):link.append(tag_a[i].get_attribute("href"))
driver.close()
```

Use `.text` to get `InnerText` attribute. Avoid using `InnerHTML`

```python
link = "https://www.bbc.com/news/articles/cd1646ewzleo"
driver = webdriver.Firefox()
driver.get(link) #opens a "firefox" window

art = driver.find_elements(By.TAG_NAME,'article') #Contents of <article>
if not art : print("Some problem with link")

headline = art[0].find_elements(By.CSS_SELECTOR,'[data-component="headline-block"]')
if not headline:print("Some problem")


arttext = [el.text for el in art]
headtext = [el.text for el in headline]
print(arttext)
print(headtext)
driver.close()
```
